```python
import cv2
import matplotlib.pyplot as plt

# Read the image
original_image = cv2.imread("sona.jpg")

# Convert the image to grayscale
grayscale_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

# Calculate the histogram for the original image
hist_original = cv2.calcHist([original_image], [0], None, [150], [0, 150])

# Calculate the histogram for the grayscale image
hist_gray = cv2.calcHist([grayscale_image], [0], None, [150], [0, 150])

# Plot the images
plt.figure(figsize=(20, 20))
plt.subplot(441), plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)), plt.axis('off'), plt.title('Original', size=20)
plt.subplot(442), plt.imshow(grayscale_image, cmap='gray'), plt.axis('off'), plt.title('Grayscale', size=20)

# Plot the histograms
plt.subplot(443), plt.hist(hist_original.ravel(), bins=256, color='blue'), plt.title('Histogram of Original Image')
plt.subplot(444), plt.hist(hist_gray.ravel(), bins=256, color='green'), plt.title('Histogram of Grayscale Image')

plt.show()
```
