```python
def count_pages(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
        num_chars = len(content)
        num_words = len(content.split())
        num_lines = content.count('\n') + 1
        num_paragraphs = len(content.split('\n\n'))
        num_pages = num_chars / 1800  # Assuming average 1800 characters per page
        
        return {
            "pages": round(num_pages),
            "lines": num_lines,
            "paragraphs": num_paragraphs,
            "words": num_words
        }

file_path = "SampleTextFile.txt"  # Change to your file path
counts = count_pages(file_path)
print("Number of Pages:", counts["pages"])
print("Number of Lines:", counts["lines"])
print("Number of Paragraphs:", counts["paragraphs"])
print("Number of Words:", counts["words"])
```

```python
def demonstrate_string_functions(file_path):
    with open(file_path, 'r') as file:
        content = file.read()

        # String functions
       
        print("\n1. Uppercase:")
        print(content.upper())
        print("\n2. Lowercase:")
        print(content.lower())
        print("\n3. Capitalize:")
        print(content.capitalize())
        print("\n4. Title:")
        print(content.title())
        print("\n5. Swapcase:")
        print(content.swapcase())
        print("\n6. Strip:")
        print(content.strip())
        print("\n7. Replace 'text' with 'string':")
        print(content.replace('text', 'string'))
        print("\n8. Split by space:")
        print(content.split())
        print("\n9. Split by lines:")
        print(content.splitlines())
        print("\n10. Startswith 'The':")
        print(content.startswith('The'))
        print("\n11. Endswith 'file':")
        print(content.endswith('file'))
        print("\n12. Find 'Python':")
        print(content.find('Python'))
        print("\n13. Count 'is':")
        print(content.count('is'))
        print("\n14. Length of text:")
        print(len(content))
        print("\n15. Is digit:")
        print(content.isdigit())
        print("\n16. Is alpha:")
        print(content.isalpha())
        print("\n17. Is alphanumeric:")
        print(content.isalnum())

file_path = "SampleTextFile.txt"  # Change to your file path
demonstrate_string_functions(file_path)
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Assuming your DataFrame has only one column
text_column_index = 0

def preprocess_text(text):
    # Tokenization
    tokens = nltk.word_tokenize(text.lower())
    
    # Removing special characters and keeping only alphanumeric tokens
    tokens = [token for token in tokens if token.isalnum()]
    
    # Removing stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    
    # Stemming
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(token) for token in tokens]
    
    return tokens

# Apply the preprocessing function to the first column
df['preprocessed_text'] = df.iloc[:, text_column_index].apply(preprocess_text)

# Display the DataFrame with the new preprocessed text column
print(df.head())
```
